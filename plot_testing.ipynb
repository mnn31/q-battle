```python
x = np.linspace(1, 1000, num=100)
print(x)
len(x)
```

```python
y = (x**(2))/(x**3/5)
y
```

```python
fig, ax = plt.subplots(1, figsize=(15,8))
ax.plot(x,y)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("YOLO")
plt.show()
```

```python
    import numpy as np
    import matplotlib.pyplot as plt

    def complex_function(z):
        return z**3 - 1 # Example complex function

    reals = np.linspace(-2, 2, 400)
    imags = np.linspace(-2, 2, 400)
    Z = reals[np.newaxis, :] + 1j * imags[:, np.newaxis]

    W = complex_function(Z)

    plt.imshow(np.angle(W), cmap='hsv', origin='lower',
               extent=[reals.min(), reals.max(), imags.min(), imags.max()])
    plt.colorbar(label='Argument')
    plt.title('Visualization of a Complex Function')
    plt.xlabel('Real Part')
    plt.ylabel('Imaginary Part')
    plt.show()
```
